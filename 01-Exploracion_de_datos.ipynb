### **1. Entropía de Shannon**

Desde la teoría se reporta que:

*La comprensión original de la entropía, que en la termodinámica se entiende como una medida de la distribución aleatoria de los elementos de un sistema que se desenvuelve en un campo de probabilidades, en la comunicación política se redimensiona, entendiéndola como una medida del ruido, de las interferencias, de la pérdida de energía de los mensajes, del autismo emisor, de la perplejidad con incertidumbres y de la incomunicación.*

Tomado de: https://erbol.com.bo/opinion/entrop%C3%ADa-comunicacional-y-pol%C3%ADtica

La entropía de Shannon nos da información de cómo es la cantidad de información que contiene un arreglo en cuestión, por lo que consideramos aplicarla en algunas columnas para determinar si la información que contienen es variada.

```python
#Ahora genero la función que me calcula la entropía de Shannon
def calcularShannon(data):

  #Ahora simplemente reciclo código que se usó en anteriores trabajos
  p , _ = np.histogram(data,bins='fd') # p permite conocer el número de veces que se encuentra la muestra, teniendo encuenta que se utiliza el bins='fd'(estimador de Freedman Diaconis) 
  p = p/len(p) # se establece la frecuencia relativa de cada muestra, la cual se puede calcular contando el número de veces que se encuentra la muestra en la serie temporal dividido por el número total de muestras.

  h = -p*np.log2(p, out=p, where=p > 0) #se calcula la entropia de shannon
  h[np.isnan(h)] = 0.0 #se reemplazan los valores vacios de h por 0
  H = np.sum(h) #se realiza la sumatoria de los valores de h
  return H #se retorna el valor de H, el cual indica la entropia de shannon
```

### **3. Prueba de normalidad**

Para la realización de esta prueba hay muchos tipos de test de hipótesis. En este caso por la cantidad de información que se posee se hará uso de la prueba de Kolmogorov-Smirnov, pues se poseen más de 300 datos. Esta prueba tiene las siguientes condiciones:


*   H0 o Hipótesis nula: Los datos siguen una distrubución normal.
*   H1 o Hipótesis alternativa: Los datos no siguen una distribución normal.

Fuente: https://github.com/thomas-haslwanter/statsintro_python/blob/master/ISP/Code_Quantlets/07_CheckNormality_CalcSamplesize/checkNormality/ISP_checkNormality.py

Para este tipo de pruebas lo normal es establecer un intervalo de confianza del 95%; en nuestro caso haremos uso de este con el objetivo de ser lo más precisos posible.

Nuestra idea inicial es que ninguno de los datos sigue una distribución normal, sin embargo, realizamos esta prueba para comprobarlo.



```python
#Desarrollo la función que me calcula el valor
def kolmogorov(data):
  _, pVals = stats.kstest((data-np.mean(data))/np.std(data,ddof=1), 'norm')
  return pVals
```

### **4. Prueba U Mann-Whitney:**

Esta prueba se aplica para distribuciones que no son paramétricas, es decir: No siguen una distribución normal en específico, por lo cual es justamente lo que se necesita.

Esta prueba tiene las siguientes condiciones:


*   H0 o Hipótesis nula: Los datos pertenecen a una misma distribución.
*   H1 o Hipótesis alternativa: Los datos pertenecen a distribuciones diferentes.

Esta prueba recibe como parámetros dos vectores, por lo cual se deben hacer agrupaciones.

Fuente: https://jmyao17.github.io/Statistics/Nonparametric_Statistical_Significance_Tests.html

Haremos uso de este test para verificar si las columnas que tienen un mismo valor de entropía de Shannon pertenecen a distribuciones distintas o si por el contrario, poseen los mismos valores.

```python
def pruebaU(d1,d2):
  _, U_test = mannwhitneyu(d1, d2)

  return U_test
```

```python
#Aplico la prueba U para icu_patients y hosp_patients
pruebaU(dataset['icu_patients'].fillna(0),dataset['hosp_patients'].fillna(0))
```

```python
#Aplico la prueba U para icu_patients y new_vaccinations
pruebaU(dataset['icu_patients'].fillna(0),dataset['new_vaccinations'].fillna(0))
```

```python
#Aplico la prueba U para new_vaccinations y hosp_patients
pruebaU(dataset['new_vaccinations'].fillna(0),dataset['hosp_patients'].fillna(0))
```

De nuevo, los valores que entrega el pVal son inferiores al 0.05, lo cual entonces nos lleva a rechazar la hipótesis nula y con ello encontramos que en efecto la información proviene de distribuciones diferentes.

El resultado de esta prueba nos da cierta tranquilidad respecto a que seguramente no hubo confusiones en el reporte de datos, pues demuestra que hay independencia entre estas columnas.

```python
def dfplot(df, country):
    #Selecting the 7 key columns for country in dataset
    df=df[df['location']==country].copy()
    df=df[['date','total_cases','new_cases',
         'total_deaths','new_deaths',
         'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred',
         'new_deaths_per_million', 'new_cases_per_million']].copy()

    #Convert to datetime
    df.date = pd.to_datetime(df['date'])

    #Fixing the mistakes in chilean data.
    if country=='Argentina':
        df.loc[df['date']=='2021-10-07','total_deaths'] = 37533
        df.loc[df['date']=='2021-10-07','new_deaths'] = 21
        df.loc[df['date']=='2021-10-08','new_deaths'] = 9
        df.loc[df['date']=='2021-10-07','new_deaths_per_million'] = 1.092
        df.loc[df['date']=='2021-10-08','new_deaths_per_million'] = 0.468
    else:
        pass

    df.set_index('date', inplace=True)
    df['7 days MA new cases'] = 0
    df['7 days MA new cases'] = df['new_cases'].rolling(7).mean()
    df['7 days MA new deaths'] = 0
    df['7 days MA new deaths'] = df['new_deaths'].rolling(7).mean()
    df['7 days MA new cases per million'] = 0
    df['7 days MA new cases per million'] = df['new_cases_per_million'].rolling(7).mean()
    df['7 days MA new deaths per million'] = 0
    df['7 days MA new deaths per million'] = df['new_deaths_per_million'].rolling(7).mean()
    df[['new_cases', '7 days MA new cases']].plot(figsize = (15, 5), alpha = 0.5)
    plt.title(f'Timeline new cases in {country}')

    df[['new_deaths', '7 days MA new deaths']].plot(figsize = (15, 5), alpha = 0.5)
    if country=='Argentina':
        plt.ylim([0,400])
    else:
        pass 


    plt.title(f'Timeline new deaths in {country}')

    df[['people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred']].plot(figsize = (15, 5), alpha = 0.5)
    plt.ylim([0,100])
    plt.fill_between(df.index, df.people_vaccinated_per_hundred)
    plt.fill_between(df.index, df.people_fully_vaccinated_per_hundred)
    plt.title(f'Timeline percentage of people vaccinated in {country}')
  
    #Return the dataframe processed
    return df
```
